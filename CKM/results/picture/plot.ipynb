```python
import matplotlib.pyplot as plt
import pandas as pd

# 
desktop_path = os.path.expanduser("/Users/cuimengmeng/Desktop")

# SVG
save_path = os.path.join(desktop_path, 'C3_fre.svg')
# 
data = """Latent: 34, Frequency: 5
Latent: 1450, Frequency: 4
Latent: 1145, Frequency: 4
Latent: 2020, Frequency: 4
Latent: 1685, Frequency: 3
Latent: 1749, Frequency: 3
Latent: 1164, Frequency: 3
Latent: 1565, Frequency: 3
Latent: 262, Frequency: 3
Latent: 1316, Frequency: 3
Latent: 201, Frequency: 2
Latent: 433, Frequency: 2
Latent: 1564, Frequency: 2
Latent: 1959, Frequency: 2
Latent: 658, Frequency: 2
Latent: 1913, Frequency: 1
Latent: 1618, Frequency: 1
Latent: 1045, Frequency: 1
Latent: 32, Frequency: 1
Latent: 1082, Frequency: 1
Latent: 570, Frequency: 1
Latent: 405, Frequency: 1
Latent: 792, Frequency: 1
Latent: 1003, Frequency: 1
Latent: 954, Frequency: 1
Latent: 321, Frequency: 1
Latent: 1258, Frequency: 1
Latent: 1908, Frequency: 1
Latent: 793, Frequency: 1"""

latent_values = []
frequencies = []

for line in data.split('\n'):
    latent, frequency = line.split(', ')
    latent_value = int(latent.split(': ')[1])
    frequency_value = int(frequency.split(': ')[1])
    
    latent_values.append(latent_value)
    frequencies.append(frequency_value)

# DataFrame
df = pd.DataFrame({
    'Latent': latent_values,
    'Frequency': frequencies
})

# Latent
df_sorted = df.sort_values(by='Latent')

# 
plt.figure(figsize=(12, 6))
bars = plt.bar(df_sorted['Latent'].astype(str), df_sorted['Frequency'], color='orange')

# 
plt.yticks(range(0, 7))  # y07
plt.grid(axis='y', linestyle='--', alpha=0.6)

# 
plt.xlabel('Latent')
plt.ylabel('Frequency')
plt.title('Latent vs Frequency')

# x
plt.xticks(rotation=90)
plt.tight_layout()
# SVG
plt.savefig(save_path, format='svg', dpi=300)
print(f"SVG file saved at: {save_path}")
```
